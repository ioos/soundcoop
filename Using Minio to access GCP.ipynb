```python
from minio import Minio
import re
from pathlib import Path
import os.path
```

```python
client = Minio('storage.googleapis.com')
bucket = 'noaa-passive-bioacoustic'
objects = list(client.list_objects(bucket, prefix='soundcoop/FRAM/FRAM_F5-17_SV1088_20160724-20170718_HMD/'))
```

```python
local_path = 'netcdf_sound_files/FRAM_F5-17_SV1088_20160724-20170718_HMD'
if(not os.path.isdir(local_path)):
    Path(local_path).mkdir()
```

```python
ct = 0
for i, obj in enumerate(objects):
    object_name = obj.object_name
    path_name = Path(object_name).name
    if(not path_name.startswith('.')):
        download_path = local_path + '/' + Path(object_name).name
        if(os.path.isfile(download_path) == True):
            print('Already downloaded: ')
            print(download_path)
        else:
            print('Download ' + str(ct) + ' of ' + str(len(objects)) + ': ' + download_path)
            ct = ct + 1
            object_data = client.get_object(bucket, object_name)
            with open(str(download_path), 'wb') as file_data:
                for data in object_data:
                    file_data.write(data)
            file_data.close()
```
